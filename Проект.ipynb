```python
from sympy.plotting import plot
from sympy import *
```

```python
# Определить корни
x = Symbol('x')
f = sin(x)**2 - cos(x)**2
solve(f, x)
```

```python
# Найти интервалы, на которых функция возрастает, убывает
diff_f = diff(f, x)
pprint(solve(diff_f > 0, x))
pprint(solve(diff_f < 0, x))
```

```python
# Построить график
gr = plot(f)
```

```python
# Вычислить вершину
solve(diff_f, x)
```

```python
# Определить промежутки, на котором f > 0
solve(f > 0, x)
```

```python
# Определить промежутки, на котором f < 0
solve(f < 0, x)
```
